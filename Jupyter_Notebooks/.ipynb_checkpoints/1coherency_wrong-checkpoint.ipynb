## Calculate Coherency of Channel i for one frequency:

### $C_{ij}=\frac{\langle x_{i} x_j^*\rangle}{\sqrt{\langle x_{i} x_{i}^*\rangle \langle x_j x_j^*\rangle}} =
\frac{\langle r_{i} r_j \rangle}{\sqrt{\langle r_{i}^2 \rangle \langle r_j^2 \rangle}} \langle e^{i\Delta \Phi} \rangle$

```python
def coherency_avg(amp, phases, freqi, el1, el2):
    
    amps1 = amp[el1, freqi]
    amps2 = amp[el2, freqi]
    
    phases1 = phases[el1, freqi]
    phases2 = phases[el2, freqi]

    r1r2_mean = np.mean(amps1 * amps2)
    
    r_sqr_mean1 = np.mean(amps1**2)
    r_sqr_mean2 = np.mean(amps2**2)
    
    sqrt = (r_sqr_mean1 * r_sqr_mean1)**(1/2)
    norm_mean = r1r2_mean / sqrt # amplitudes r

    dphi = phases1 - phases2
    exp = np.exp(1j * dphi) # e^i*dphi
    exp_mean = np.mean(exp)
    
    return norm_mean * exp_mean
```

## Coherence $= |C_{ij}|$

## Imaginary Part of Coherency $= Im\{C_{ij}\}$

## Calculate coherence and imaginary part of coherency

```python
print(phases_on.shape, amp_on.shape)
```

```python
def coherence_all(amp, phases, freqi):
    
    N = phases.shape[0]    

    # Initialize
    coherence_mat = np.zeros([N, N])
    imcohy_mat = np.zeros([N, N])

    for comb in combinations_with_replacement(range(N), 2):
        
        coherency = coherency_avg(amp, phases, freqi, *comb)
        coherence = abs(coherency)
        imcohy = np.imag(coherency)
        abs_imcohy = abs(imcohy)
        
        coherence_mat[comb] = coherence
        imcohy_mat[comb] = abs_imcohy
        
        # Mirror for nicer symmetric plot
        coherence_mat[comb[1], comb[0]] = coherence
        imcohy_mat[comb[1], comb[0]] = abs_imcohy
    
    return coherence_mat, imcohy_mat
```

```python
coherence_on = np.zeros([amp_on.shape[1], amp_on.shape[0], amp_on.shape[0]])
imcohy_on = np.zeros([amp_on.shape[1], amp_on.shape[0], amp_on.shape[0]])

coherence_off = np.zeros([amp_off.shape[1], amp_off.shape[0], amp_off.shape[0]])
imcohy_off = np.zeros([amp_off.shape[1], amp_off.shape[0], amp_off.shape[0]])

for freqi in range(amp_on.shape[1]):
    coherence_on[freqi], imcohy_on[freqi] = coherence_all(amp_on, phases_on, freqi)
    
for freqi in range(amp_off.shape[1]):
    coherence_off[freqi], imcohy_off[freqi] = coherence_all(amp_off, phases_off, freqi)
```

```python
max_imcohy = np.max([np.max(imcohy_on), np.max(imcohy_off)])
min_imcohy = np.min([np.min(imcohy_on), np.min(imcohy_off)])

max_imcohy_diff = np.max(imcohy_on - imcohy_off)
min_imcohy_diff = np.min(imcohy_on - imcohy_off)

max_coh_diff = np.max(coherence_on - coherence_off)
min_coh_diff = np.min(coherence_on - coherence_off)
```

```python
for freqi in [13, 17]:

    fig, ax = plt.subplots(2, 3, figsize=(15, 7.5))
    
    im1 = ax[0, 0].imshow(coherence_on[freqi])
    im2 = ax[1, 0].imshow(imcohy_on[freqi])
    im3 = ax[0, 1].imshow(coherence_off[freqi])
    im4 = ax[1, 1].imshow(imcohy_off[freqi])
    im5 = ax[0, 2].imshow(coherence_on[freqi] - coherence_off[freqi])
    im6 = ax[1, 2].imshow(imcohy_on[freqi] - imcohy_off[freqi])
    
    cb1 = plt.colorbar(im1, ax=ax[0, 0])
    cb2 = plt.colorbar(im2, ax=ax[1, 0])
    cb3 = plt.colorbar(im3, ax=ax[0, 1])
    cb4 = plt.colorbar(im4, ax=ax[1, 1])
    cb5 = plt.colorbar(im5, ax=ax[0, 2])
    cb6 = plt.colorbar(im6, ax=ax[1, 2])
    
    im1.set_clim(vmin=0, vmax=1)
    im2.set_clim(vmin=min_imcohy, vmax=max_imcohy)
    im3.set_clim(vmin=0, vmax=1)
    im4.set_clim(vmin=min_imcohy, vmax=max_imcohy)
    im5.set_clim(vmin=min_coh_diff, vmax=max_coh_diff)
    im6.set_clim(vmin=min_imcohy_diff, vmax=max_imcohy_diff)
    
    cb1.set_label("Coherence")
    cb2.set_label("Absolute Imcohy")
    cb3.set_label("Coherence")
    cb4.set_label("Absolute Imcohy")
    cb5.set_label("Coherence Diff")
    cb6.set_label("Absolute Imcohy Diff")

    ax[0, 0].set_title("On")
    ax[1, 0].set_title("On")
    ax[0, 1].set_title("Off")
    ax[1, 1].set_title("Off")
    ax[0, 2].set_title("On - Off")
    ax[1, 2].set_title("On - Off")
    
    for axes in ax.flatten():
        axes.set_xticks(np.arange(amp_on.shape[0]))
        axes.set_yticks(np.arange(amp_on.shape[0]))
        axes.set_xticklabels(voxel_nms, rotation=45)
        axes.set_yticklabels(voxel_nms)

    
    plt.suptitle(f"Frequency: {freqs[freqi]:.0f} Hz", position=(0.5, 1.03))
    plt.tight_layout()
    plt.savefig(f"/Users/moritzgerster/Documents/Code/Litvak11/plots/Coherence{freqs[freqi]:.0f}Hz.pdf", bbox_inches="tight")
    plt.show()
```
