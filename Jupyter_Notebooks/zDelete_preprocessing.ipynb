# With filter:

```python
filt_freqs = np.arange(50, 101, 50)

raw_filt_on = raw_on.copy().notch_filter(freqs=filt_freqs, picks=picks, notch_widths=.1)
raw_filt_off = raw_off.copy().notch_filter(freqs=filt_freqs, picks=picks, notch_widths=.1)
```

```python
# Calculate power spectra across the the continuous data
sample_r = 2400
nperseg = 1*sample_r
spectra_filt_on, freqs_filt_on = psd_welch(raw_filt_on, picks=picks, n_overlap=nperseg // 2., n_fft=nperseg, n_per_seg=nperseg, average="median")
spectra_filt_off, freqs_filt_off = psd_welch(raw_filt_off, picks=picks, n_overlap=nperseg // 2., n_fft=nperseg, n_per_seg=nperseg, average="median")
```

```python
fm = FOOOF(peak_width_limits=(2, 10), max_n_peaks=5, min_peak_height=0, peak_threshold=2, aperiodic_mode="knee", verbose=True)

fig, ax = plt.subplots(3, 2, figsize=(13, 13), sharex=True, sharey=True)
ax = ax.flatten()
for i in range(6):
    fm.fit(freqs_filt_on, spectra_filt_on[i], freq_range)
    fm.plot(ax=ax[i])
   # fm.print_results()
    ax[i].set_title(picks[i], fontsize=20)
    if i < 8:
        ax[i].get_legend().remove()
plt.tight_layout()
fig.suptitle("On", fontsize=50, position=(0.5, 1.05))
plt.show()
        
fig, ax = plt.subplots(3, 2, figsize=(13, 13), sharex=True, sharey=True)
ax = ax.flatten()
for i in range(6):
    fm.fit(freqs_filt_off, spectra_filt_off[i], freq_range)
    fm.plot(ax=ax[i])
   # fm.print_results()
    ax[i].set_title(picks[i], fontsize=20)
    if i < 8:
        ax[i].get_legend().remove()
fig.suptitle("Off", fontsize=50, position=(0.5, 1.05))
plt.tight_layout()
plt.show()
```

#### ON STN

```python
fg_on_stn = FOOOFGroup(peak_width_limits=(2, 10), max_n_peaks=5, min_peak_height=0, peak_threshold=2, aperiodic_mode="knee", verbose=True)
fg_on_stn.fit(freqs_filt_off, spectra_filt_off[3:9], freq_range)
fg_on_stn.plot()
```

#### Off STN

```python
fg_off_stn = FOOOFGroup(peak_width_limits=(2, 10), max_n_peaks=5, min_peak_height=0, peak_threshold=2, aperiodic_mode="knee", verbose=True)
fg_off_stn.fit(freqs_filt_off, spectra_filt_off[3:9], freq_range)
fg_off_stn.plot()
```

```python
#exp_on_ctx = fg_on_ctx.get_params('aperiodic_params', 'exponent')
exp_on_stn = fg_on_stn.get_params('aperiodic_params', 'exponent')
#exp_off_ctx = fg_off_ctx.get_params('aperiodic_params', 'exponent')
exp_off_stn = fg_off_stn.get_params('aperiodic_params', 'exponent')
```
